#### <u> Example 1 </u>

```python
import numpy as np

# Inisialisasi input dan target
inputs = [
    (np.array([1, 1, 1]), 1),
    (np.array([1, 0, 1]), -1),
    (np.array([0, 1, 1]), -1),
    (np.array([0, 0, 1]), -1)
]


# Inisialisasi bobot dalam persentase (50% untuk tiga fitur)
weights = np.array([0.5, 0.5, 0.5])  # Bobot awal 50% untuk tiap fitur
bias = 0  # Bias diatur ke 0 untuk sederhana

# Threshold untuk menentukan output
threshold = 0.8

# Learning rate
learning_rate = 0.1

# Fungsi aktivasi menggunakan threshold
def activation(weighted_sum):
    return 1 if weighted_sum >= threshold else -1

# Training loop
epoch = 0
while epoch < 2:  # Batasi hingga 2 epoch
    print(f"Epoch {epoch+1}")
    no_update = True  # Untuk mengecek apakah ada perubahan pada epoch ini
    
    for i, (input_vec, target) in enumerate(inputs):
        # Menghitung weighted sum (z = w·x + b)
        weighted_sum = np.dot(weights, input_vec) + bias
        output = activation(weighted_sum)

        # Tentukan aksi sesuai dengan tabel
        action = "None"
        if output != target:
            action = "Decrease" if target == -1 else "Increase"
            # Update bobot sesuai action
            weights -= learning_rate * input_vec
            bias -= learning_rate * target
            no_update = False  # Ada pembaruan pada epoch ini

        # Menampilkan hasil perhitungan untuk setiap input
        print(f"Input: {input_vec}, Weight: {weights * 100}%, "
              f"Weighted Sum: {weighted_sum}, Output Produced: {output}, "
              f"Target: {target}, Action: {action}")

    print("\n")
    epoch += 1

    # Jika tidak ada pembaruan, hentikan loop lebih awal
    if no_update:
        break

print("Training selesai.")
```

#### <u> Example 2 </u>

```python
import numpy as np

# Inisialisasi input dan target
inputs = [
    (np.array([1, 0, 0]), 0),
    (np.array([0, 0, 1]), 0),
    (np.array([0, 1, 1]), 0),
    (np.array([1, 1, 1]), 1)
]

# Inisialisasi bobot (50%) untuk tiga fitur
weights = np.array([0.5, 0.5, 0.5])  # Bobot awal 50% untuk tiap fitur
bias = 0  # Bias diatur ke 0 untuk sederhana

# Learning rate
learning_rate = 0.1

# Threshold
threshold = 0.8

# Fungsi aktivasi menggunakan formula dari gambar
def activation(y_in, theta):
    if y_in > theta:
        return 1
    elif -theta <= y_in <= theta:
        return 0
    else:
        return -1

# Training loop
epoch = 0
while epoch < 2:  # Loop maksimal 2 epoch
    print(f"Epoch {epoch+1}")
    no_update = True  # Untuk mengecek apakah ada perubahan pada epoch ini
    
    for i, (input_vec, target) in enumerate(inputs):
        # Menghitung y_in (z = w·x + b)
        y_in = np.dot(weights, input_vec) + bias
        output = activation(y_in, threshold)

        # Tentukan apakah ada perubahan bobot dan bias
        action = "None"
        if output < target:  # Jika output lebih kecil dari target
            action = "Increase"
            weights += learning_rate * input_vec  # Tingkatkan bobot
            bias += learning_rate * target  # Update bias
            no_update = False  # Ada perubahan
        elif output > target:  # Jika output lebih besar dari target
            action = "Decrease"
            weights -= learning_rate * input_vec  # Kurangi bobot
            bias -= learning_rate * target  # Update bias
            no_update = False  # Ada perubahan

        # Menampilkan hasil perhitungan untuk setiap input
        print(f"Input: {input_vec}, Weight: {weights * 100}%, "
              f"y_in: {y_in:.1f}, Output Produced: {output}, "
              f"Target: {target}, Action: {action}")

    print("\n")
    epoch += 1

    # Jika tidak ada pembaruan pada epoch ini, hentikan loop
    if no_update:
        break

print("Training selesai.")
```
